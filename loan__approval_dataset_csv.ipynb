```python
df.columns = df.columns.str.strip()
```

```python
# Featue and target separation
x = df.drop("loan_status",axis=1)
y = df["loan_status"]
print(x.columns)
# in this
# y is what we want to predict
# x is the data after we remove the target to aviod data leakage
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(
x,y,
test_size=0.2,
random_state=42,
stratify=y
)
```

```python
if 'loan_id' in x_train.columns:
    x_train = x_train.drop("loan_id",axis=1)
if 'loan_id' in x_test.columns:
    x_test = x_test.drop("loan_id", axis=1)
# loan id is just an identifier which has no meaning , if we k
# it,the model may learn noise
# removing unnecessary colums from the model train and test
```

```python
x_train.dtypes
```

```python
#encoding the model x train adn x test
# one hot encode categorical features

from sklearn.preprocessing import OneHotEncoder

categorical_cols = ["education","self_employed"]
one = OneHotEncoder(drop="first",sparse_output=False)
x_train_cat = one.fit_transform(x_train[categorical_cols])
x_test_cat = one.transform(x_test[categorical_cols])

# the drop is to remove the redundant columns to dummy varibale trap
# sparse in default would store them as numpy array,
# if false itll give the encoded data as a normal table of
#numbers insted compressing it
```

```python
# Scaling numeric fetures

from sklearn.preprocessing import StandardScaler
numeric_cols = x_train.drop(columns=categorical_cols).columns
scaler = StandardScaler()
x_train_num = scaler.fit_transform(x_train[numeric_cols])
x_test_num = scaler.transform(x_test[numeric_cols])
```

```python
x_train_num_df = pd.DataFrame(
x_train_num,
columns=numeric_cols,
index=x_train.index
)

x_train_num_df.head()

#x_test_num_df=pd.DataFrame(
    #x_test_num,
    #columns=numeric_cols,
    #index=x_test.index
    #)
```

```python
# feature combination or concatenation
x_train_final = np.hstack([x_train_num,x_train_cat])
x_test_final = np.hstack([x_test_num,x_test_cat])
```

```python
# here logistic regression is the corrent one
#because the target is categorical output - approved/not approved
#logistic regression is used for classification and perdicts pr
# converts it into class 0 or 1, works well as baseline Model
#its simple, fast and easy to implement
```

```python
from collections import UserString
#logistic regression model

from sklearn.linear_model import LogisticRegression
model=LogisticRegression(max_iter=1000)
model.fit(x_train_final,y_train)

#max iter = 1000 is for logistic regression to leant using
#the number of iterations(optimization)
#default iteration are not enough model fails to cover
#this is where the model learns the pattern from data
```

```python
# evaluate accuracy
from sklearn.metrics import accuracy_score

y_pred = model.predict(x_test_final)
accuracy_score(y_test,y_pred)
```
